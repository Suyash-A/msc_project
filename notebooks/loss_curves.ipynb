```python
import matplotlib.pyplot as plt
import pandas as pd

# Read the output.txt file
data = pd.read_csv('../data_msc_project/bertdata/batch64-please4-bluebert-balancedset-chexblabeled/loss.txt', header=None, names=['epoch', 'iterations', 'loss'], index_col=False)
data = data.drop('iterations', axis=1)

data['epoch'] = range(1, 289)
data['epoch'] = data['epoch'] / 36

# # Plot only the loss against index
# plt.figure(figsize=(10, 6))
# plt.plot(data['epoch'], data['loss'])
# plt.grid()
# # plt.yscale('log')
# plt.xlabel('Epoch')
# plt.ylabel('Loss')
# plt.title('Loss Curve')
# plt.show()

# Create a figure with two subplots side by side
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Plot on the first subplot with normal y-scale
axes[0].plot(data['epoch'], data['loss'])
axes[0].set_title('Loss Curve (Linear Scale)')
axes[0].set_xlabel('Epoch')
axes[0].set_ylabel('Loss')
axes[0].grid()
axes[0].legend()

# Plot on the second subplot with log y-scale
axes[1].plot(data['epoch'], data['loss'])
axes[1].set_yscale('log')
axes[1].set_title('Loss Curve (Log Scale)')
axes[1].set_xlabel('Epoch')
axes[1].set_ylabel('Loss')
axes[1].grid()
axes[1].legend()

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Read the first output.txt file
data1 = pd.read_csv('../data_msc_project/bertdata/batch64-please4-bluebert-balancedset-chexblabeled/loss.txt', header=None, names=['epoch', 'iterations', 'loss'], index_col=False)
data1 = data1.drop('iterations', axis=1)

data1['epoch'] = range(1, 289)
data1['epoch'] = data1['epoch'] / 36

# Read the second output.txt file
data2 = pd.read_csv('../data_msc_project/bertdata/batch64-please4-bluebert-chexset-chexblabeled/loss.txt', header=None, names=['epoch', 'iterations', 'loss'], index_col=False)
data2 = data2.drop('iterations', axis=1)

data2['epoch'] = range(1, 137)
data2['epoch'] = data2['epoch'] / 17

# Plot both losses on the same graph
plt.figure(figsize=(10, 6))

# Plot the first loss curve
plt.plot(data1['epoch'], data1['loss'], label='Balanced Dataset')

# Plot the second loss curve
plt.plot(data2['epoch'], data2['loss'], label='CheXpert Dataset')

plt.yscale('log')

# Add labels and title
plt.grid()
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Loss Curves')

# Add a legend to differentiate the lines
plt.legend()

# Show the plot
plt.show()
```
